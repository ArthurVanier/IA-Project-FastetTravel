```python
def read_csv_time ( path):
    
    find_id = False
    find_trajet = False 
    depart = []
    arrive = []
    durees = []

    try:
        with open(path, 'r') as f:

            for line in f :

                duree = ""
                trajet = ""

                for cara in line:

                    if cara.isdigit() and find_id :
                        find_trajet = True
                        duree += cara
                    if find_id and not find_trajet and cara != "\t":
                        trajet += cara
                    if cara == "\t" and not find_id :
                        find_id = True
                    if cara == '\n':
                        pass
                
                find_id = False
                find_trajet = False 

                trajet = trajet.split(" - ")
                depart += [trajet[0]]
                arrive += [trajet[1]]

                durees+=[int(duree)]
    except IOError as e:
        print('Une erreur est survenue lors de la lecture du fichier :', e)

    return pd.DataFrame({'depart': depart, 'arrivee': arrive ,'duree': durees})
```

```python
df_travels = read_csv_time('timetables.csv')
df_travels
df_travels.to_csv('timetables_clear.csv',  sep=";" , index=False , encoding='utf-8')
```
