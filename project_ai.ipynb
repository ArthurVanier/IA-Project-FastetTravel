```python
def read_csv_time ( path):
    
    find_id = False
    find_trajet = False 
    depart = []
    arrive = []
    durees = []

    try:
        with open(path, 'r') as f:

            for line in f :

                duree = ""
                trajet = ""

                for cara in line:

                    if cara.isdigit() and find_id :
                        find_trajet = True
                        duree += cara
                    if find_id and not find_trajet and cara != "\t":
                        trajet += cara
                    if cara == "\t" and not find_id :
                        find_id = True
                    if cara == '\n':
                        pass
                
                find_id = False
                find_trajet = False 

                trajet = trajet.split(" - ")
                depart += [trajet[0]]
                arrive += [trajet[1]]

                durees+=[int(duree)]
    except IOError as e:
        print('Une erreur est survenue lors de la lecture du fichier :', e)

    return pd.DataFrame({'depart': depart, 'arrivee': arrive ,'duree': durees})
```

```python
import csv

def extract_unique_stations(input_csv, output_csv):
    unique_stations = set()

    # Lire le fichier CSV et extraire les noms des gares
    with open(input_csv, 'r', encoding='utf-8') as file:
        reader = csv.reader(file, delimiter=';')
        for row in reader:
            unique_stations.add(row[0])  # Gare de départ
            unique_stations.add(row[1])  # Gare d'arrivée

    # Écrire les noms uniques des gares dans un nouveau fichier CSV
    with open(output_csv, 'w', encoding='utf-8', newline='') as file:
        writer = csv.writer(file)
        for station in sorted(unique_stations):
            writer.writerow([station])

# Appel de la fonction
extract_unique_stations('timetables_clear.csv', 'gares_uniques.csv')
```

```python
import spacy
nlp = spacy.load('fr_core_news_sm')
df_gares = pd.read_csv('gares_uniques.csv')
gares = df_gares['gare'].to_string()
print(gares)
text = "Je mange beaucoup de Paris et j'aimerais en manger plus. Je voudrias arriver à Marseille en partant de paris et acheter des pommes à Amiens. "
doc = nlp(text)

```

```python
identified_gares = [ent.text for ent in doc.ents if ent.text in gares]
print("Gares identifiées :", identified_gares)
```
